# Gradient Boosting Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df.head()
```

```python
X = df.drop('class', axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50, 100], 'learning_rate':[0.1, 0.05, 0.2], 'max_depth':[3, 4, 5]}
```

```python
gb_model = GradientBoostingClassifier()
```

```python
grid = GridSearchCV(gb_model, param_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
predicitions = grid.predict(X_test)
```

```python
predicitions
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
print(classification_report(y_test, predicitions))
```

```python
# grid.best_estimator_.feature_importances_
```

```python
feat_import = grid.best_estimator_.feature_importances_
```

```python
imp_feat = pd.DataFrame(index=X.columns, data=feat_import, columns=['Importance'])
```

```python
imp_feat = imp_feat[imp_feat['Importance']>0.0005]
```

```python
imp_feat = imp_feat.sort_values("Importance")
```

```python
plt.figure(figsize=(14,8), dpi=250)
sns.barplot(data=imp_feat, x=imp_feat.index, y='Importance')
plt.xticks(rotation=90);
```
