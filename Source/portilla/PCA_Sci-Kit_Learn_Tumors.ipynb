# Principal Component Analysis Sci-Kit Learn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
# help(PCA)
```

```python
pca_model = PCA(n_components=8)
```

```python
pca_results = pca_model.fit_transform(scaled_X)
```

```python
pca_results
```

```python
plt.figure(figsize=(10,10), dpi=250)
plt.scatter(pca_results[:,0], pca_results[:,1])
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
can_dic = load_breast_cancer()
```

```python
plt.figure(figsize=(10,10), dpi=250)
plt.scatter(pca_results[:,0], pca_results[:,1], c=can_dic['target'])
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8',], columns=df.columns)
```

```python
plt.figure(figsize=(20, 3), dpi=150)
sns.heatmap(df_comp, annot=True)
```

```python
pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1, 30):
    pca=PCA(n_components=n)
    pca.fit(scaled_X)

    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(10, 10), dpi=150)
plt.plot(range(1,30), explained_variance)
plt.xlabel('Number of Components ')
plt.ylabel('Variance Explained')
```
