```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values , eigen_vestors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vestors
```

```python
eigen_vestors.shape
```

```python
# How Many Principal Components Do We Want
num_components = 2
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
sorted_key
```

```python
eigen_values, eigen_vestors = eigen_values[sorted_key], eigen_vestors[:,sorted_key]
```

```python
eigen_values
```

```python
principal_components = np.dot(scaled_X, eigen_vestors)
```

```python
principal_components
```
